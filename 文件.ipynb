显示目录下的所有文件和子目录：

```julia
for (root, dirs, files) in walkdir(".")
    println("正在显示$(root)内容---子目录：")
    for dir in dirs
        println(joinpath(root, dir)) # path to directories
    end
    println("正在显示$(root)内容---文件：")
    for file in files
        println(joinpath(root, file)) # path to files
    end
end
```
